```python
LotArray = ["LOT", "LOT:", "LOT.", "LO", "LO.", "LO:"] 
ExpArray = ["EXP", "EXP.", "EXP:", "MFG", "MFG.", "MFG:", "USE BY", "USE BY.", "USE BY:", "BEST BEFORE",
            "BEST BEFORE.", "BEST BEFORE:", "MEILLEUR AVANT", "MEILLEUR AVANT.", "MEILLEUR AVANT:"] 
MonthArray = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC", "JA", "FE", 
              "MR", "AL", "MA", "JN", "JL", "AU", "SE", "OC", "NO", "DE"]
DateSep = ["-", ".", "/", ""]
```

```python
BigArray = LotArray + ExpArray + MonthArray
```

```python
import random
import time
for exp in range(len(ExpArray)):
    for mon in range(len(MonthArray)):
        for day in range(1,32):
            random.seed(time.clock())
            year = random.randint(1900,2100)
            TwoDigitYear = random.randint(0,99)
            BigArray.append(ExpArray[exp] + " " + str(year) + " " + MonthArray[mon] + " " + str(day))
            BigArray.append(ExpArray[exp] + " " + str(TwoDigitYear).zfill(2) + " " + MonthArray[mon] + " " + str(day))
            BigArray.append(ExpArray[exp] + " " + MonthArray[mon] + " " + str(day) + " " + str(year)) 
```

```python
for exp in range(len(ExpArray)):
    for mon in range(1,13):
        for day in range(1,32):
            for sep in range(len(DateSep)):
                random.seed(time.clock())
                year = random.randint(1900,2100)
                TwoDigitYear = random.randint(0,99)
                BigArray.append(ExpArray[exp] + " " + str(year) + DateSep[sep] + str(mon).zfill(2) + DateSep[sep] + str(day).zfill(2))
                BigArray.append(ExpArray[exp] + " " + str(mon).zfill(2) + DateSep[sep] + str(day).zfill(2) + DateSep[sep] + str(year))
                BigArray.append(ExpArray[exp] + " " + str(day).zfill(2) + DateSep[sep] + str(mon).zfill(2) + DateSep[sep] + str(year))
                BigArray.append(ExpArray[exp] + " " + str(day).zfill(2) + DateSep[sep] + str(mon).zfill(2) + DateSep[sep] + str(TwoDigitYear).zfill(2))
```

```python
import random
import string
BigLotArray = []
for lot in range(len(LotArray)):
    for counter in range(5000):
        random.seed(time.clock())
        TwoDigitNumber = random.randint(0,99)
        ThreeDigitNumber = random.randint(0,999)
        Letter1 = random.choice(string.ascii_uppercase)
        Letter2 = random.choice(string.ascii_uppercase)
        BigLotArray.append(LotArray[lot] + " " + str(TwoDigitNumber).zfill(2) + "-" + str(ThreeDigitNumber).zfill(3) + "-" + Letter1 + Letter2)
```
